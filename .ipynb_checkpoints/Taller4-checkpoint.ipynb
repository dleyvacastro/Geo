```python
import sympy as sp
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def norm(x):
    return sp.sqrt(np.dot(x,x)).simplify()
def simplify(x):
    return [sp.simplify(i) for i in x]
def diff(x, var, degree=1):
    return [sp.diff(i, var, degree) for i in x]

def show(x,row=True):
    if row:
        display(sp.Matrix([x]))
    else:
        display(sp.Matrix(x))

sp.init_printing()
```

# Punto 2

```python
theta, phi, a, b, c, t = sp.symbols('theta, phi, a, b, c, t', reals=True)
X = (a*sp.sin(theta)*sp.cos(phi), b*sp.sin(theta)*sp.sin(phi), c*sp.cos(theta))
dX_theta = diff(X, theta)
dX_ttheta = diff(X,theta,2)
dX_phi = diff(X, phi)
dX_pphi = diff(X,phi,2)
dX_thetaphi = diff(X, theta, phi)
```

```python
display(sp.Matrix([X]))
display(sp.Matrix([dX_theta]))
display(sp.Matrix([dX_ttheta]))
display(sp.Matrix([dX_phi]))
display(sp.Matrix([dX_pphi]))
display(sp.Matrix([dX_thetaphi]))
```

## Primera forma fundamental

```python
E = np.dot(dX_theta,dX_theta).simplify()
F = np.dot(dX_theta,dX_phi).simplify()
G = np.dot(dX_phi,dX_phi).simplify()
```

```python
show(E)
show(F)
show(G)
```

## Tensor Metrico

```python
g = sp.Matrix([[E,F],[F,G]])
g
```

## Elemento de Area

```python
Ae = norm(np.cross(dX_theta,dX_phi))
# Ae.subs(theta, np.pi).subs(phi,np.pi)
```

```python
Ae = Ae.simplify()
```

```python
Ae
```

## Area de la superficie

```python
# lambdyfy the function
Ae = Ae.subs({a:3, b:2, c:1})
f = sp.lambdify((theta, phi), Ae)
```

```python
integrate.dblquad(f, 0, 2*np.pi, lambda x: 0, lambda x: np.pi)
```

## Segunda forma fundamental

```python
N = np.cross(dX_theta, dX_phi)/norm(np.cross(dX_phi, dX_theta))
N = simplify(N)
display(N)
print(N)
```

```python
e = np.dot(N, dX_ttheta).simplify()
f = np.dot(N, dX_thetaphi).simplify()
g = np.dot(N, dX_pphi).simplify()
```

```python
display(e)
display(f)
display(g)
```

## Curvaturas

```python
K = ((e*g-f**2)/(E*G-F**2)).simplify() # Gaussian c.
K
```

```python
H = (0.5*(E*g-2*F*f+G*e)/(E*G-F**2)).simplify() # Mean c.
H
```

```python
k2 = (H + sp.sqrt(H**2-K))
k1 = (H - sp.sqrt(H**2-K))
display(k1)
display(k2)
```

## Vector Normal

```python
display(N)
```

```python
p = (np.pi/2,np.pi/2)
Np = [i.subs({theta:p[0], phi:p[1]}) for i in N]
Np
```

## Plot

```python
u = np.linspace(0 , 2 * np.pi , 100 )
v = np.linspace(0 , 2*np.pi , 100 )
u, v = np.meshgrid(u , v )
a0 = 3
b0 = 2
c0 = 1

d = a0*np.sin(u)*np.cos(v)
e = b0*np.sin(u)*np.sin(v)
g = c0*np.cos(u)

dX_theta0 = [i.subs({a: a0, b:b0, c:c0, theta:p[0], phi:p[1]}) for i in dX_theta]
dX_phi0 = [i.subs({a: a0, b:b0, c:c0, theta:p[0], phi:p[1]}) for i in dX_phi]


x_p = X[0].subs({a: 3, b:2, c:1, theta: np.pi/2, phi: np.pi/2})
y_p = X[1].subs({a: 3, b:2, c:1, theta: np.pi/2, phi: np.pi/2})
z_p = X[2].subs({a: 3, b:2, c:1, theta: np.pi/2, phi: np.pi/2})

Np = [i.subs({a: a0, b:b0, c:c0}) for i in Np]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlim([-2.5, 2.5])
ax.plot_surface(d, e, g, alpha = 0.5)
ax.scatter(x_p, y_p, z_p, color='r')
ax.view_init(azim=10)

ax.quiver(x_p, y_p, z_p, *dX_theta0, color = (1, 0, 0, 1))
ax.quiver(x_p, y_p, z_p, *dX_phi0, color = (0, 1, 0, 1))
ax.quiver(x_p, y_p, z_p, *Np, color = (0, 0, 1, 1))
```

# Punto 3

```python
X = ((1+0.5*t*sp.cos(theta/2))*sp.cos(theta)), ((1+0.5*t*sp.cos(theta/2))*sp.sin(theta)), 0.5*t*sp.sin(theta/2)
dX_theta = diff(X, theta)
dX_ttheta = diff(X,theta,2)
dX_t = diff(X, t)
dX_t2 = diff(X,t,2)
dX_thetat = diff(X, theta, t)
```

```python
display(sp.Matrix([X]))
display(sp.Matrix([dX_theta]))
display(sp.Matrix([dX_ttheta]))
display(sp.Matrix([dX_t]))
display(sp.Matrix([dX_t2]))
display(sp.Matrix([dX_thetat]))
```

## Primera forma fundamental

```python
E = np.dot(dX_theta,dX_theta).simplify()
F = np.dot(dX_theta,dX_t).simplify()
G = np.dot(dX_t,dX_t).simplify()
```

```python
show(E)
show(F)
show(G)
```

## Tensor Metrico

```python
g = sp.Matrix([[E,F],[F,G]])
g
```

## Elemento de Area

```python
Ae = norm(np.cross(dX_theta,dX_t))
Ae
```

## Area de la superficie

```python
# lambdyfy the function
f = sp.lambdify((theta, t), Ae)
```

```python
integrate.dblquad(f, -0.25, 0.25,0, 2*np.pi)
```

## Vector Normal

```python
N = np.cross(dX_theta, dX_t)/norm(np.cross(dX_t, dX_theta))
# N = simplify(N)
display(N)
print(N)
```

```python
N0 = [i.subs(t,0) for i in N]

display([i.subs(theta,0) for i in N0])
display([i.subs(theta,2*np.pi) for i in N0])
```

```python
fig = plt.figure(figsize = (9, 5))
ax = plt.axes(projection = "3d")
a = np.linspace(0 , 2.0*np.pi , 30 )
b = np.linspace(-0.25 , 0.25 , 5 )
a , b = np.meshgrid(a,b)
x = ( 1 + 0.5 * b * np . cos ( a/2.0 ) ) * np.cos( a )
y = ( 1 + 0.5 * b * np . cos ( a/2.0 ) ) * np.sin( a )

z = 0.5 * b * np.sin( a/2.0)
ax.plot_surface(x, y, z, linewidth = 2, alpha = 0.8)
ax.set_xlabel( "X" )
ax.set_ylabel( "Y" )
ax.set_zlabel( "Z" )

# plot N0
ax.quiver(1, 0, 0, *[i.subs(theta,0) for i in N0], color = (1, 0, 0, 1), length = 0.3, normalize = True, label = r"$N_0$ at $\theta = 0$")
ax.quiver(1, 0, 0, *[i.subs(theta,2*np.pi) for i in N0], color = (0, 1, 0, 1), length = 0.3, normalize = True, label = r"$N_0$ at $\theta = 2\pi$")

ax.set_title( "Mobius strip" , weight = "bold" )
ax.set_zlim([-0.3, 0.3])
plt.legend()
plt.show()
```

Vea que la cinta de Möbius es una superficie no orientable, por que no tiene un vector normal bien definido, teniendo 2 direcciones distintas cuando esta teoricamente en un mismo punto, como se puede ver en la figura.
